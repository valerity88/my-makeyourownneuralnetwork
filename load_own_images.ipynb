```python
# помогает загрузить данные из файла изображения PNG 
import imageio

# помогает выбрать несколько файлов используя шаблоны
import glob

# импортируем numpy
import numpy

# библиотека для построения массивов
import matplotlib.pyplot

# обеспечивает построение графиков внутри этого блокнота, 
# а не во внешнем окне
%matplotlib inline



# наш собственный набор тестовых данных изображения
our_own_dataset = []

for image_file_name in glob.glob('my_own_images/2828_my_own_?.png'):
    print("loading ... ", image_file_name)
    # используем имя файла для установления корректной метки
    label = int(image_file_name[-5:-4])
    # загружаем данные из png файла внутрь массива
    img_array = imageio.imread(image_file_name, as_gray = True)
    # изменяем из 28x28 на список из 784 значений, инвертируем значения
    img_data = 255.0 - img_array.reshape(784)
    # затем масштабируем данные в диапазон от 0,01 до 1,0
    img_data = (img_data / 255.0 * 0.99) + 0.01
    print(numpy.min(img_data))
    print(numpy.max(img_data))
    # добавляем метку и данные изображения в тестовый набор данных
    record = numpy.append(label, img_data)
    print(record)
    our_own_dataset.append(record)
    pass
    
```

```python
matplotlib.pyplot.imshow(our_own_dataset[3][1:].reshape(28,28),\
                         cmap = 'Greys', interpolation = 'None')
```
